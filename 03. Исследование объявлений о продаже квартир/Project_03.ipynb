Разберемся сначала с пропусками в столбце с количеством балконов.

````python
# <заменим пропуски в столбце с количеством балконов на 0>
df['balcony'] = df['balcony'].fillna(0)

# <проверим наличие пропусков в столбце с количеством балконов и выведем его методом isnull()>
df['balcony'].isnull().sum()
````

````python
# <заменим пропуски в столбце с высотой потолков на 0>
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())

# <проверим наличие пропусков в столбце с высотой потолков и выведем его методом isnull()>
df['ceiling_height'].isnull().sum()
````

Пропущенные значения по высоте потолков в квартирах, выставленных на продажу, заполнены медианным значением по солбцу – это значение близко к реальному среднему, так как медиана не чувствительна к выбросам (если они есть).

Займемся пропусками в столбце с жилой площадью.

````python
# <рассчитаем коэффициент соотношения жилой площади к общей и добавим в таблицу эти данные в новый столбец>
df['coefficient_living_area'] = df['living_area'] / df['total_area']

# <заменим пропуски в столбце с жилой площадью на 0>
df['living_area'] = df['living_area'].fillna(0)

# <создадим Data Frame, содержащий в себе количество комнат из таблицы 
# и медианные значения коэффициентов соотношения жилой площади к общей соответственно>
median_living_area_coefficient = df.groupby('rooms').agg({'coefficient_living_area': 'median'})

# <создадим функцию, которая заменят нулевые значения в столбце c жилой площадью на медианные, 
# в зависимости от количества комнат в квартире>
def living_area(row):
    rooms = row['rooms']
    if row['living_area'] == 0:
        row['living_area'] = row['total_area'] * median_living_area_coefficient.loc[rooms, 'coefficient_living_area']
    return row

# <применим полученую функцию к столбцу 'living_area'>
df = df.apply(living_area, axis = 1)

# <проверим наличие пропусков в столбце с жилой площадью и выведем его методом isnull()>
df['living_area'].isnull().sum()
````

Пропуски в столбце с жилой площадью заменены на медианные значения, в зависимости от количества комнат.

Аналогично поступим и с пропусками в информации по площади кухни, с учетом того, что в квартирах-студиях (количество комнат – *0*) кухни совмещены с комнатой.

````python
# <рассчитаем коэффициент соотношения площади кухни к общей и добавим в таблицу эти данные в новый столбец>
df['coefficient_kitchen_area'] = df['kitchen_area'] / df['total_area']

# <заменим пропуски в столбце с площадью кухни на 0>
df['kitchen_area'] = df['kitchen_area'].fillna(0)

# <заменим пропуски в столбце с коэффициентами соотношения площади кухни к общей на 0, 
# т. к. в квартирах-студиях кухня совмещена комнатой>
df['coefficient_kitchen_area'] = df['coefficient_kitchen_area'].fillna(0)

# <создадим функцию, которая заменят нулевые значения в столбце c площадью кухни на медианные, 
# в зависимости от количества комнат в квартире>
median_kitchen_area_coefficient = df.groupby('rooms').agg({'coefficient_kitchen_area': 'median'})

def kitchen_area(row):
    rooms = row['rooms']
    if row['kitchen_area'] == 0:
        row['kitchen_area'] = row['total_area'] * median_kitchen_area_coefficient.loc[rooms, 'coefficient_kitchen_area']
    return row

# <применим полученую функцию к столбцу 'kitchen_area'> 
df = df.apply(kitchen_area, axis = 1)

# <проверим наличие пропусков в столбце с площадью кухни и выведем его методом isnull()>
df['kitchen_area'].isnull().sum()
````

И пропуски в столбце с площадью кухни – обработаны.

Посмотрим, что можно сделать с пропусками в столбцах с информацией по количеству этажей в доме и населённому пункту.
