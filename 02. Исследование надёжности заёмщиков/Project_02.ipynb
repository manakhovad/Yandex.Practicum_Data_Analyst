### Обработка пропусков

Проверим данные на наличие пропусков вызовом набора методов для суммирования пропущенных значений.

```python
# <суммарное количество пропусков, выявленных методом isnull() в таблице df>
df.isnull().sum()
```

Пустые значения присутствуют только в солбцах *days_employed* и *total_income*. Тип данных, хранящийся в обоих столбцах — *float64*.

Стоит заметить, что количество пропущенных значений в столбце *days_employed* соответствует количеству пропущенных значений в столбце *total_income*.

Пропуски в данных столбцах могли возникнуть по причине того, что клиенты банка не указали данную информацию о себе.

Пропущенная информация, содержащаяся в столбцах *days_employed* (трудовой стаж в днях) и *total_income* (доход в месяц), не оказывает влияния на зависимость между способностью заёмщика погасить кредит в срок и его семейным положением и количеством детей. Но удалять строки с пропущенными значениями нельзя, так как мы потеряем большое количество данных, содержащих информацию о семейном положении и наличии детей у заёмщиков. Поэтому пропущенные значения в столбцах *days_employed* (трудовой стаж в днях) и *total_income* (доход в месяц) заменим на медианные значения для каждого из типов занятости.

Определим количество типов занятости клиентов банка, представленных в таблице.

```python
# <посчитаем количество типов занятости клиентов банка методом .count() в сгруппированной по столбцу 'income_type' таблице 
#методом .groupby()>
df.groupby('income_type').count()
```

Заменим пропущенные значения в столбце *total_income* (доход в месяц) на медианные по каждому из типов занятости (безработный, в декрете, госслужащий, компаньон, пенсионер, предприниматель, сотрудник, студент).

```python
# Создадим список с типами занятасти клиентов банка
income_type_list = ['безработный', 'в декрете', 'госслужащий', 'компаньон', 'пенсионер', 'предприниматель', 'сотрудник', 'студент']

# Рассчитаем медианные значения ежемесячного дохода для каждого из типов занятости клиентов банка
for a in income_type_list:
    income_type = df.loc[df['income_type'] == a]
    median_total_income_type = income_type['total_income'].median()
    print('Медиана дохода в месяц для категории', a, 'составляет', median_total_income_type)
    print()
    
# Заменим пропущенные значения в столбце total_income (доход в месяц) на медианные по каждому из типов занятости
    df.loc[df['income_type'] == a, 'total_income'] = df.loc[df['income_type'] == a, 'total_income'].fillna(median_total_income_type)
print()    
display(df.head(15))
```

Заменим пропущенные значения в столбце *days_employed* (трудовой стаж в днях) на медианные по каждому из типов занятости (безработный, в декрете, госслужащий, компаньон, пенсионер, предприниматель, сотрудник, студент).

```python
# Рассчитаем медианные значения трудового стажа для каждого из типов занятости клиентов банка
for a in income_type_list:
    income_type = df.loc[df['income_type'] == a]
    median_days_employed_income_type = income_type['days_employed'].median()
    print('Медиана трудового стажа категории', a, 'составляет', median_days_employed_income_type)
    print()
    
# Заменим пропущенные значения в столбце days_employed (трудовой стаж в днях) на медианные по каждому из типов занятости
    df.loc[df['income_type'] == a, 'days_employed'] = df.loc[df['income_type'] == a, 'days_employed'].fillna(median_days_employed_income_type)
print()    
display(df.head(15))
```

Проверим данные на наличие пропусков вызовом набора методов для суммирования пропущенных значений.

```python
# <суммарное количество пропусков, выявленных методом isnull() в таблице df>
df.isnull().sum()
```

### Лемматизация

В столбце с целями получения кредита (*purpose*) приведена информация, предоставленная клиентами банка. Внимательно посмотрев на первые несколько строк, становится понятно, что одинаковые цели получения кредита сформулированы и записаны по-разному.
Напрмер:
* *на проведение свадьбы* и *сыграть свадьбу*;
* *покупка жилья* и *операции с жильем*;
* и т. д.

Для более удобного восприятия информации и упрощения проведения анализа информации сгруппируем похожие цели займа по категориям.

Для определения наиболее востребованных целей получения кредита лемматизируем столбец с целями получения кредита (*purpose*) и определим ключевые слова в целях займа.

```python
# <Импортируем библиотеку pymystem3>
from pymystem3 import Mystem
m = Mystem()

# <Лемматизируем слова в столбце 'purpose' методом m.lemmatize и добавим их в столбец 'purpose_lemma'>
# <Столбец 'purpose_lemma' добавим в таблицу методом .apply()>
df['purpose_lemma'] = df['purpose'].apply(m.lemmatize)

# <Подсчитаем сколько раз каждое слово встречается в столбце 'purpose_lemma' и выведем результат>
from collections import Counter
print(Counter(df['purpose_lemma'].sum()))
```

Как ы видим, наиболее часто встречающимися словами, определющими цель займа являются *'недвижимость'*, *'жилье'*, *'автомобиль'*, *'образование'*, *'свадьба'*.

Сгруппируем цели получения кредита следующим образом:
* все указанные цели, которые связаны с покупкой жилой или коммерческой недвижимости запишем в группу *Займ на покупку недвижимости*;
* все цели, которые связаны с получением образования (доплнительного или основного) — *Займ на образование*;
* цели на приобретение транспортных средств — *Займ на покупку автомобиля*;
* цели на проведение свадьбы — *Займ на проведение свадьбы*.

Полученные категории целей получения кредита запишем в столбец *'purpose_category'*.

```python
# <Создадим функцию, которая будет перебирать значения столбца 'purpose_lemma', икать в них слова-маркеры
# каждой категории займов и возвращать соответствующую категорию>
def purpose_category(lemmas):
    if ('жилье' in lemmas) or ('недвижимость' in lemmas):
        return 'Займ на покупку недвижимости'
    if 'образование' in lemmas:
        return 'Займ на образование'
    if ('сыграть' in lemmas) or ('свадьба' in lemmas):
        return 'Займ на проведение свадьбы'
    if 'автомобиль' in lemmas:
        return 'Займ на покупку автомобиля'

# <Добавим методом .apply() в таблицу столбец 'purpose_category', содержащий категории целей займов>
df['purpose_category'] = df['purpose_lemma'].apply(purpose_category)

# <Выведем первые 5 строк на экран>
display(df.head())
```

* покупка недвижимости;
* обазование;
* покупка автомобиля;
* проведение свадьбы.

>**_<span style="color:green">Красивое решение через общий список всех целей и применением счётчика Counter. Оценив общую частотность и встречаемость лемм в нашем датасете, ты смогла правильно выделить общие категории.</span>_**

```python
# <Методом .abs() исправим все "минусы" в столбце 'children' на "плюсы">
df_cat['children'] = df_cat['children'].abs().copy()

# <Проверим столбец с количеством детей на наличие отрицательных значений>
df_cat['children'].value_counts()
```

Согласно типологии семей по количеству детей выделяют такие разновидности семей:
* бездетные (или инфертильные);
* малодетные (семьи с одним-двумя детьми);
* многодетные (трое и более детей).

Категоризируем информацию о заёмщиках по типологии семей по количеству детей.

```python
# <Создадим функцию, которая возвращает категорию семьи по количеству детей children, используя правила:
# - 'бездетные' при значении children = 0;
# - 'малодетные' при значении children = 1 или 2;
# - 'многодетные' во всех остальных случаях>
def children_category(children):

        if children == 0:
                return 'бездетные'
        if children == 1 or children == 2:
                return 'малодетные'
        return 'многодетные'

# <Добавим в таблицу столбец с категориями семей по количеству детей и проверим сколько заёмщиков относятся к каждой группе>
df_cat['children_category'] = df_cat['children'].apply(children_category)
df_cat.groupby('children_category').count()
```

Сравнив полученный результат с предыдущей проверкой, видим, что столбец с количеством детей не содержит аномальных значений и выдает требуемый результат при обработке функцией, возвращающей категорию семьи по количеству детей.

```python
# <Добавим в таблицу df_cat столбец с категорией клиента по уровню дохода ('income_category').
# Применим для этого метод .qcut() к столбцу 'total_income' и поделим содержащиеся в нём значения на три равные категории:
# - с низким уровнем дохода;
# - со средним уровнем дохода;
# - с высоким уровнем дохода.>
df_cat['income_category'] = pd.qcut(df['total_income'],3, ['низкий', 'средний', 'высокий'])

# <Выделим таблицу, содержащую информацию только о задолженностях по кредиту,
# уровнем дохода клиентов банка и соответствующей категорией по уровню дохода>
df_cat_income = df_cat[['debt','total_income','income_category']]

# <Создадим сводную таблицу по категориям семей в зависимости от ежемесячного дохода,
# содеращую в столбцах информацию о наличии задолженностей по кредитам>
income_debt_pivot = df_cat_income.pivot_table(index='income_category', columns='debt', aggfunc='count')

# <Переименуем столбцы для удобства восприятия информации>
income_debt_pivot.columns = ['нет долга', 'есть долг']
income_debt_pivot
```

Кредиты на проведение свадьбы гасятся чуть хуже, скорее всего в силу молодого возраста заёмщиков, но наличие супруга / супруги с доходом позволяет гасить кредиты в срок.

Просрочки, связанные с выплатами по кредитам на образование, могут быть вызваны дефицитом времени клиентов для совмещения работы и обучения, а соответственно и дефицитом средств на выплату кредита.

Наиболее часто просрочки по кредитам встречаются в категории на покупку автомобилей. Данная категория займов достаточно рискованная для банка, поэтому сопровождается дополнительными расходами клиентов на страховки, а так же имеет достаточно высокие процентные ставки. Эта категория - одна из самых "дорогих" на рынке кредитных услуг.
